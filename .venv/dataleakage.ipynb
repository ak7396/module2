```python
#TASK3
"Task 3 — Compute Baseline MAE Build a baseline model that predicts the mean"
" selling_price for every record. Calculate and "
"print the MAE of this baseline on the full cleaned dataset."
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
# Step 1: Calculate mean selling price
mean_price = df['selling_price'].mean()

# Step 2: Create baseline predictions (same value for all rows)
y_true = df['selling_price']
y_pred = [mean_price] * len(df)

# Step 3: Calculate MAE
mae = mean_absolute_error(y_true, y_pred)

# Step 4: Print results
print("Mean Selling Price (Baseline):", mean_price)
print("Baseline MAE:", mae)
```
